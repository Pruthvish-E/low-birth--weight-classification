```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import metrics
from warnings import filterwarnings
filterwarnings("ignore")
%matplotlib inline
data = pd.read_csv("./cleaned.csv")
df = data
```

```python
feature_cols = ['community', 'age', 'weight1', 'history','HB','IFA','BP1', 'res']
X = data[feature_cols] 
y1 = data['reslt']
```

```python
#from medium
def svm_function(x,y):
    #initilizing weight
    w = np.zeros(len(x[0]))
    #initialize learning rate
    l_rate = 1
    #epoch
    epoch = 10
    #output list
    out = []
    #training svm
    for e in range(epoch):
        for i, val in enumerate(x):
            val1 = np.dot(x[i], w)
            if (y[i]*val1 < 1):
                w = w + l_rate * ((y[i]*x[i]) - (2*(1/epoch)*w))
            else:
                w = w + l_rate * (-2*(1/epoch)*w)
    
    for i, val in enumerate(x):
        out.append(np.dot(x[i], w))
    
    return w
```

```python
from sklearn.model_selection import KFold
kf = KFold(5, True)
kf.get_n_splits(X)
averageSVM = 0
 
for i in range(10):
    for train_index, test_index in kf.split(X):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y1.iloc[train_index], y1.iloc[test_index]
        
        x = np.array(X_train[['weight1']].values.tolist())
        y2 = y_train.values.tolist()
        
        t = []
        for i in y2:
            if(i == 0):
                t.append(-1)
            else:
                t.append(1)
        y = np.array(t)
        
        w = svm_function(x,y)
        x_t = np.array(X_test[['weight1']].values.tolist())
        result = []
        for i, val in enumerate(x_t):
                result.append(np.dot(x_t[i], w))
        y_pred = []
        for i in result:
            if(i<0):
                y_pred.append(-1)
            else:
                y_pred.append(1)
                
        averageSVM+=metrics.accuracy_score(y_test, pd.DataFrame(y_pred))
        #print(metrics.accuracy_score(y_test, pd.DataFrame(y_pred)))
        
        

        
        
print("Average Accuracy simple:- ",averageSVM/50)

    
```
